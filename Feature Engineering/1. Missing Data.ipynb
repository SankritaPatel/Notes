Mean Imputation Technique
- Mean Value Imputation: THis technique is work when data is normally distributed.

```python
sns.distplot(df['age'])
```

```python
df.age.isnull().sum()
```

```python
df['age_mean'] = df['age'].fillna(df['age'].mean())
```

```python
df[['age_mean', 'age']]
```

- Median Value Imputation: If you have outliers in dataset.

```python
df['age_median'] = df['age'].fillna(df['age'].median())
```

```python
df[['age_median', 'age']]
```

- Mode Value Imputation: This is used mostly for categorical data.

```python
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
df[df['age'].notna()]
```

```python
mode = df[df['age'].notna()]['embarked'].mode()[0]
```

```python
mode
```

```python
df['embarked_fill'] = df['embarked'].fillna(mode)
```

```python
df[['embarked_fill', 'embarked']]
```

- Random Sample Imputation: Select any Random Value and put it into NA.
